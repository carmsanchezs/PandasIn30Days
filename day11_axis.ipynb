## Day 2. How do I use the "axis" parameter in pandas?

```python
import pandas as pd 
import ssl
```

```python
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.head()
```

```python
drinks.drop('continent', axis=1).head()  # the change is not in place
```

```python
drinks.drop(2, axis=0).head() 
```

```python
drinks.drop([1,2,3,4,5], axis=0).head() 
```

```python
drinks.drop('country', axis='columns').head() 
```

```python
drinks.drop([4,5], axis='index').head() 
```
