## How do i sort a pandas Data Frame or a Series?

```python
import pandas as pd 
import ssl
```

```python
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

New way to sort, new versions

```python
# sort a Series, it is not changing the DataFrame
movies.title.sort_values(ascending=False)  # order in alphabetic order, numbers first
```

```python
# order the DataFrame by a Series
movies.sort_values('title')
```

```python
movies.sort_values('duration', ascending=False)  # the longest movies in the dataset
```

How to sort by multiple columns

```python
movies.sort_values(['genre', 'duration'], ascending=[True, False])
```

```python
# it doesn't change the original dataFrame 'movies', unless you set inplace=True
movies.head()
```
