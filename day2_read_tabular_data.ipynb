## Part of the day 10.
Reading from a csv file only two columns and ignore others

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.columns
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports', usecols=['City', 'State'])
ufo.columns
```

```python
ufo.head()
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports', usecols=[0, 3])
ufo.columns
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports', nrows=5)  # most useful for big datasets
ufo
```
